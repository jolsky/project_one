```python
wrkdf.iloc[:, 7] = wrkdf.iloc[:, 7].replace("no disponible", 0)
wrkdf.iloc[:, 7] = wrkdf.iloc[:, 7].astype(float)
wrkdf.iloc[:, 7].fillna(wrkdf.iloc[:, 7].mean(), inplace=True)

wrkdf.iloc[:, 8] = wrkdf.iloc[:, 8].replace("no disponible", 0)
wrkdf.iloc[:, 8] = wrkdf.iloc[:, 8].astype(float)
wrkdf.iloc[:, 8].fillna(wrkdf.iloc[:, 8].mean(), inplace=True)

wrkdf.iloc[:, 9] = wrkdf.iloc[:, 9].replace("no disponible", 0)
wrkdf.iloc[:, 9] = wrkdf.iloc[:, 9].astype(float)
wrkdf.iloc[:, 9].fillna(wrkdf.iloc[:, 9].mean(), inplace=True)
```

```python
wrkdf.columns = wrkdf.columns.str.lower()

wrkdf.rename(columns={"fecha_evento": "fecha",
                      "tipo_da„o": "temptipo",
                      "detalle_da„o": "detalle",
                      "costo_original_pre": "cst_ol_pre",
                      "monto_total_de_atenciîn_pre": "mto_tot_at_pre",
                      "costo_original_cien": "cst_ol_cien", 
                      "monto_total_de_atenciîn_cien": "mto_tot_at_cien",}, inplace=True)
```

```python
wrkdf["detalle"] = wrkdf["detalle"].str.lower()
```

```python
wrkdf["detalle"] = wrkdf["detalle"].str.replace("reparaciîn", "reparacion")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da„os menores", "danos menores")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da„o severo", "dano severo")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da–o menor", "dano menor")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da„os severos", "danos severos")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da„os exteriores", "danos exteriores")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("da„os en recubrimientos", "danos en recubrimientos")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("hidrçulicas", "hidraulicas")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("elƒctricas", "electricas")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("reconstrucci—n parcial", "reconstruccion parcial")
wrkdf["detalle"] = wrkdf["detalle"].str.replace("informaci—n en construcci—n", "informacion en construccion")
```

```python
wrkdf[["tipo_num", "tipo"]] = wrkdf["temptipo"].str.split(" ", n=1, expand=True)

wrkdf.drop(columns=["temptipo"], inplace=True)
```

```python
wrkdf.iloc[:, 11].fillna(3, inplace=True)
wrkdf.iloc[:, 11] = wrkdf.iloc[:, 11].astype(int)
```

```python
wrkdf[["d_a","d_b","d_c","d_d","d_e","d_f"]] = wrkdf["detalle"].str.split(";", n=5, expand=True)
wrkdf.drop(columns=["detalle"], inplace=True)
```

```python
wrkdf["d_a"] = wrkdf["d_a"].str.replace("1.-", "")
wrkdf["d_b"] = wrkdf["d_b"].str.replace("2.-", "")
wrkdf["d_c"] = wrkdf["d_c"].str.replace("3.-", "")
wrkdf["d_d"] = wrkdf["d_d"].str.replace("4.-", "")
wrkdf["d_e"] = wrkdf["d_e"].str.replace("5.-", "")
wrkdf["d_f"] = wrkdf["d_f"].str.replace("6.-", "")
```

```python
wrkdf["d_a"] = wrkdf["d_a"].str.strip()
wrkdf["d_b"] = wrkdf["d_b"].str.strip()
wrkdf["d_c"] = wrkdf["d_c"].str.strip()
wrkdf["d_d"] = wrkdf["d_d"].str.strip()
wrkdf["d_e"] = wrkdf["d_e"].str.strip()
wrkdf["d_f"] = wrkdf["d_f"].str.strip()
```
